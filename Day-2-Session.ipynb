```python
import pandas as pd
u_cols = ['user_id', 'gender', 'age', 'occupation', 'zip_code']  #UserID::Gender::Age::Occupation::Zip-code
users = pd.read_csv('users.dat', sep='::', names=u_cols)

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp'] # UserID::MovieID::Rating::Timestamp
ratings = pd.read_csv('ratings.dat', sep='::', names=r_cols)

# the movies file contains columns indicating the movie's genres
# let's only load the first five columns of the file with usecols
m_cols = ['movie_id', 'title', 'genres']  #MovieID::Title::Genres

movies = pd.read_csv('movies.dat', sep='::', names=m_cols, usecols=range(5))

# create one merged DataFrame
movie_ratings = pd.merge(movies, ratings)
lens = pd.merge(movie_ratings, users)
lens.head()
```

```python
most_rated = lens.groupby('title').size().sort_values(ascending=False)[:10]
most_rated
```

```python
most_rated = lens.groupby('genres').size().sort_values(ascending=False)[:10]
most_rated
```

```python
import numpy as np
movie_stats = lens.groupby('title').agg({'rating': [np.size, np.mean]})
movie_stats.head()
```
